```python
import numpy as np
import pandas as pd
import pickle5 as pickle
```

```python
df_phrases_path = './df_phrases.pkl'
df_phrases = pd.read_pickle(df_phrases_path)
print(df_phrases.shape)
df_phrases.head(3)
```

```python
df_sentences_path = './df_sentences.pkl'
df_sentences = pd.read_pickle(df_sentences_path)
print(df_sentences.shape)
df_sentences.head(3)
```

```python
targets = np.array([ np.array(value[0]) for value in df_phrases[["embedding"]].values ])

print(f"targets:{targets.shape}")
```

```python
source = np.array(df_phrases.at[44, 'embedding'])

cosine = np.dot(targets,source)/(np.linalg.norm(targets, axis=1)*np.linalg.norm(source))
print("Cosine Similarity:\n", cosine)
```

```python
source = np.array(df_phrases.at[45, 'embedding'])

cosine = np.dot(targets,source)/(np.linalg.norm(targets, axis=1)*np.linalg.norm(source))
print("Cosine Similarity:\n", cosine)
```

```python
df_cosines = pd.DataFrame(columns=['id', 'line'])

for i, row in df_sentences.iterrows():
    line = f'{row["line"]:03}'
    # print(f'Calculating cosines for [ { row["id"] } / { line } ] {row["sentence"][:50]}...')
    source = np.array(row["embedding"])
    cosine = np.dot(targets,source)/(np.linalg.norm(targets, axis=1)*np.linalg.norm(source))
    # print("Cosine Similarity:\n", cosine)
    # Create new row
    new_row = dict([(f"Cosine{f'{key:02}'}", value) for key, value in enumerate(cosine.flatten(), 1)])
    new_row["id"] = row["id"]
    new_row["line"] = row["line"]
    df_cosines = df_cosines.append(new_row, ignore_index=True)

df_cosines
```

```python
print(df_sentences.shape)
print(df_cosines.shape)
df_comparison = pd.concat([df_sentences, df_cosines.loc[:, 'Cosine01':'Cosine64']], axis="columns")
print(df_comparison.shape)
df_comparison.head()
```

```python
threshold = 0.85
```

```python
cosine_cols = [col for col in df_comparison.columns if 'Cosine' in col]
df_comparison[(df_comparison.filter(regex='Cosine') > threshold).any(axis=1)]
```

```python
df_results = pd.DataFrame(columns=['id_transcript', 'line', 'sentence', 'phrase', 'category', 'label', 'similarity'])
unique_rows = set()

for i, row in df_comparison.iterrows():
  # print(f'Calculating sentences for [{ i }]...')
  for n in range(1,64+1):
    col = f"Cosine{f'{n:02}'}"
    if row[col] > threshold:
      unique_rows.add(i);
      # Get phrase
      phrase = df_phrases.loc[[ n - 1 ]]
      new_row = { 
        'id_transcript': row["id"],
        'line': row["line"],
        'sentence': row["sentence"],
        'phrase': df_phrases.at[n-1,"example"],
        'category': df_phrases.at[n-1,"category"],
        'label': df_phrases.at[n-1,"label"],
        'similarity': row[col]
      }
      df_results = df_results.append(new_row, ignore_index=True)

unique_rows = sorted(unique_rows)
print(unique_rows)
df_results
```

```python
df_summary = df_results.groupby(['id_transcript','label'])['similarity'].agg('max')
df_summary
```
